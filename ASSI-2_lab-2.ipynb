```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/titanic.csv")
print("Statistical simmary:")
print(df.describe())
```

```python
sns.set_style("whitegrid")
df.hist(figsize=(10, 6), bins=30, edgecolor='red')
plt.suptitle("Histograms of Dataset Features")
plt.show()

```

```python
if not df.select_dtypes(include=['object']).empty:
    cat_col = df.select_dtypes(include=['object']).columns[0]
    df[cat_col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {cat_col}")
    plt.xlabel(cat_col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
if not df.select_dtypes(include=['object']).empty:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[cat_col], palette="pastel")
    plt.title(f"Count Plot of {cat_col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
num_cols = df.select_dtypes(include=['number']).columns
if len(num_cols) > 1:
    plt.figure(figsize=(10, 6))
    plt.plot(df[num_cols[0]], df[num_cols[1]], marker="o", linestyle="-", color="b")
    plt.title(f"Line Chart of {num_cols[0]} vs {num_cols[1]}")
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df)
plt.title("Violin Plot of Numerical Features")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pair Plot of Dataset", y=1.02)
plt.show()
```
